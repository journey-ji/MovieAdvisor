# 아이템기반 최근접 이웃 필터링

## 데이터 로드하기 및 정리

```python
import pandas as pd
import numpy as np

movies = pd.read_csv('../../data/ml-latest-small/movies.csv')
ratings = pd.read_csv('../../data/ml-latest-small/ratings.csv')
print(movies.shape)
print(ratings.shape)
```

```python
ratings.head()
```

### 사용자 평점 데이터를 사용자-아이템 평점 행렬로 변환

```python

ratings = ratings[['userId','movieId','rating']]
ratings_matrix = ratings.pivot_table('rating',index='userId',columns = 'movieId')
ratings_matrix.head(3)
```

### movies 와 ratings 데이터셋을 join하기

두 데이터셋을 movieId 속성을 기준으로 조인해준다.

사용자-아이템 행렬에서 movieId 대신 title을 넣어서 가독성을 높여주기 위한 기반작업

```python
rating_movies = pd.merge(ratings,movies,on='movieId')
rating_movies.head(3)
```

### title을 컬럼으로 pivot table 생성하기

```python
ratings_matrix = rating_movies.pivot_table('rating',index='userId',columns='title')


## 위에서 평점데이터가 없어서 Nan이었던 부분을 0으로 바꿔줌
ratings_matrix = ratings_matrix.fillna(0)
ratings_matrix.head(3)
```

### 사용자-아이템 행렬을 아이템-사용자 행렬로 변환

그냥 transpose() 한번 써주면 끝

```python
ratings_matrix_T = ratings_matrix.transpose()
ratings_matrix_T.head(3)
```

## 영화와 영화들 간 유사도 산출

```python
from sklearn.metrics.pairwise import cosine_similarity
item_sim = cosine_similarity(ratings_matrix_T, ratings_matrix_T)


## 행과 열을 columns 명으로 잡기
item_sim_df = pd.DataFrame(data=item_sim, index=ratings_matrix.columns, columns = ratings_matrix.columns)
print(item_sim_df.shape)
item_sim_df.head(3)
```

```python

## 갓파더와 이름이 유사한 상위 6개 아이템 추출
item_sim_df['Godfather, The (1972)'].sort_values(ascending=False)[:6]
```

## 아이템 기반 인접 이웃 협업 필터링을 통해 개인화 영화 추천

```python
### 사용자-아이템 평점 행렬과 아이템간 유사도 행렬을 내적하여서
### 사용자의 아이템별 예측 평점 행렬을 계산한다.
def predict_rating(ratings_arr,item_sim_arr):
    ## 아래는 가중치 공식 Weidhted Rating Sum
    ratings_pred = ratings_arr.dot(item_sim_arr)/ np.array([np.abs(item_sim_arr).sum(axis=1)])
    return ratings_pred
```

```python
ratings_matrix.head(3)
```

```python
ratings_pred = predict_rating(ratings_matrix.values, item_sim_df.values)
ratings_pred_matrix = pd.DataFrame(data=ratings_pred, index =ratings_matrix.index,
                                   columns = ratings_matrix.columns)
ratings_pred_matrix.head(3)
```

## 예측성능 평가하기

```python
from sklearn.metrics import mean_squared_error

def get_mse(pred,actual):
    ## 0이 아니라 Nan이 었던 평점들에 대해서 제외하고 mse 를 구하기 위해서 아래 nonzero 메서드사용
    pred = pred[actual.nonzero()].flatten()
    actual = actual[actual.nonzero()].flatten()
    return mean_squared_error(pred,actual)

print(get_mse(ratings_pred, ratings_matrix.values))
```

### top-n 개만 유사도 기반의 예측 평점 계산하기

```python
def predict_rating_topsim(ratings_arr,item_sim_arr, n=20):
    pred = np.zeros(ratings_arr.shape)
    
    for col in range(ratings_arr.shape[1]):
        
        ## 자기자신 빼고, 내림차순으로 상위 N개의 데이터에 대해 "인덱스"를 저장함
        top_n_items = [np.argsort(item_sim_arr[:,col])[:n-1:-1]]
        
        
        ## 예측 평점 계산
        for row in range(ratings_arr.shape[0]):
            ## 곱계산
            pred[row,col] = item_sim_arr[col,:][top_n_items].dot(ratings_arr[row,:][top_n_items].T)
            ## 나누기 계산
            pred[row,col] /= np.sum(np.abs(item_sim_arr[col,:][top_n_items]))
            
    return pred
```

### top-n 개의 예측 평점에 대한 mse 계산하기

```python
### 시간이 오래걸리는 작업 ###

### !!!!!!!!!!!!!!!!! ###
ratings_pred = predict_rating_topsim(ratings_matrix.values,item_sim_df.values, n = 20)
print('아이템 기반 인점 top-20 이웃 mse: ',get_mse(ratings_pred,ratings_matrix.values))

ratings_pred_matrix = pd.DataFrame(data=ratings_pred, index= ratings_matrix.index,
                                  columns = ratings_matrix.columns)
```

## 사용자가 관람하지 않은 영화 중에서 아이템 기반 인접 이웃 협업 필터링으로 영화 추천하기

```python
def get_unseen_movies(ratings_matrix,userId):
    user_rating = ratings_matrix.loc[userId,:]
    
    already_seen = user_rating[user_rating>0].index.tolist()
    movies_list = ratings_matrix.columns.tolist()
    unseen_list = [movie for movie in movies_list if movie not in already_seen]
    
    return unseen_list
```

### 아이템 기반 유사도로 평점이 부여된 데이터 세트에서 해당 사용자가 관람하지 않은 영화들의 예측 평점이 가장 높은 영화를 추천

```python
def recomm_movie_by_userId(pred_df,userId,unseen_list,top_n=10):
    recomm_movies = pred_df.loc[userId,unseen_list].sort_values(ascending=False)[:top_n]
    return recomm_movies

unseen_list = get_unseen_movies(ratings_matrix,9)

recomm_movies = recomm_movie_by_userId(ratings_pred_matrix,9,unseen_list , top_n=10)

recomm_movies = pd.DataFrame(data=recomm_movies.values, index= recomm_movies.index,
                            columns = ['pred_score'])
recomm_movies
```
